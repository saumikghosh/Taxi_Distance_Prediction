```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv(r"C:\Users\Soumik Ghosh\Desktop\hackathon\Taxi trip\Train.csv")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data.head(2)
```

```python
x = data[['fare_estimate','tip_amount','tolls_amount','passenger_count']]
y = data['trip_distance_miles']
```

```python
x_test,x_train,y_test,y_train=train_test_split(x,y,train_size=0.2,random_state=42)
```

```python
x_test.shape,x_train.shape,y_test.shape,y_train.shape
```

```python
lin_reg = LinearRegression()
```

```python
lin_reg.fit(x_train,y_train)
```

```python
lin_reg.predict(x_test)
```
